```python
import random as rand
from math import sqrt
```

```python
def vector_subtract(v,w):
    """ subtracts corresponding elements of two vectors """
    return [v_i - w_i for v_i,w_i in zip(v,w)]

def sum_of_squares(v):
    squares = [v_i ** 2 for v_i in v]
    print(squares)
    return sum(squares)

def magnitude(v):
    # length of a vector in projection
    # basically get the hypotenuse
    print(sqrt(sum_of_squares(v)))
    return sqrt(sum_of_squares(v))

def distance(v,w):
    """ distance between two vectors """
    #print(v)
    #print(w)
    #print(vector_subtract(v,w))
    return magnitude(vector_subtract(v,w))

def step(v,direction,step_size):
    #print(v)
    #print('direction: {}'.format(str(direction)))
    return [v_i + step_size * direction_i
        for v_i, direction_i in zip(v,direction)]

def sum_of_squares_gradient(v):
    return [2 * v_i for v_i in v]
```

```python
# non-stochastic
v = [rand.randint(-10,10) for i in range(4)]
print(len(v))
tolerance = 0.00001

iters = 0
while True:
    gradient = sum_of_squares_gradient(v)
    #print('gradient: {}'.format(str(gradient)))
    next_v = step(v,gradient, -0.01)
    if distance(next_v,v) < tolerance:
        break
    v = next_v
    iters +=1
    print('==========')
print(iters)
```

```python
# stochastic
x = [rand.randint(0,10) for i in range(10)]
tolerance = 0.0000001

iters = 0
while True:
    stochIdxs = [rand.randrange(10) for i in range(10)]
    v = [x[i] for i in stochIdxs]
    gradient = sum_of_squares_gradient(v)
    
    next_v = step(v,gradient, -0.01)
    if distance(next_v,v) < tolerance:
        break
    v = next_v
    iters += 1
print(iters)
```

```python
# From calculation, it is expected that the local minimum occurs at x=9/4

cur_x = 2 # The algorithm starts at x=6
gamma = 0.01 # step size multiplier
precision = 0.00000001
previous_step_size = cur_x

def df(x):
    return 2*x**2

while previous_step_size > precision:
    prev_x = cur_x
    cur_x += -gamma * df(prev_x)
    previous_step_size = abs(cur_x - prev_x)

print("The local minimum occurs at %f" % cur_x)
```
